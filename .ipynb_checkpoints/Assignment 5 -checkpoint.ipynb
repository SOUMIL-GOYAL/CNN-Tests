## Part 2: Use ResNet50 on the project data

We have seen how ResNet50 is able to classify individual images in the previous two questions. Now let's use it for our image dataset. 

As we mentioned before, ResNet50 has learned features for many objects. In transfer learning, we do not modify most of the features(aka weights and bias) that ResNet50 already has. We simply transfer the learned features over, then leave a few layers open at the end to learn the features specific to our dataset. 

Due to machine limitation, we have trained the ResNet50 for our dataset, and will be loading the the trained model later. Running it in Codio may cause machine to freeze.

## Please do not run the following cell below in Codio. It is for observational purpose only.

```python

# create base model 
base = ResNet50(weights='imagenet', input_shape=(224, 224, 3), include_top=False)

#output classes 
class_list =  [0,1]

#build model based on ResNet50
x = base.output
x = Flatten()(x)
fc1 = Dense(1024, activation='relu')(x) 
fc2 = Dense(1024, activation='relu')(fc1) 
x = Dropout(0.5)(fc2)
predictions = Dense(len(class_list), activation='softmax')(x) 
model = Model(inputs=base.input, outputs=predictions)

#freeze the layer weights in pre-trained ResNet50
for layer in base.layers:
  layer.trainable = False

#trainning
batch_size = 256
total_images = 11811

adam = Adam(lr=0.00001)
model.compile(adam, loss='categorical_crossentropy', metrics=['accuracy'])

#save the weights 
checkpoint_path = "./cp_2.ckpt" 
checkpoint_dir = os.path.dirname(checkpoint_path)

# Create checkpoint callback
cp_callback = ModelCheckpoint(checkpoint_path, 
                                                 save_weights_only=True,
                                                 verbose=1)

history = model.fit(x_train, y_train,epochs= 5, steps_per_epoch=total_images // batch_size, shuffle=True,callbacks = [cp_callback], workers = 8)


# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.show()

```

## Part 3: Let's test our model

Please run the cells in part 3. Let's load the weights that we have trained to the model and observe it's performance. The last two cells may take a few mintues.

````python
# build the model
base = ResNet50(weights='imagenet', input_shape=(224, 224, 3), include_top=False)
class_list = [0,1]
x = base.output
x = Flatten()(x)
fc1 = Dense(1024, activation='relu')(x) 
fc2 = Dense(1024, activation='relu')(fc1) 
x = Dropout(0.5)(fc2)
predictions = Dense(len(class_list), activation='softmax')(x) 
model = Model(inputs=base.input, outputs=predictions)
adam = Adam(lr=0.00001)
model.compile(adam, loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
````

````python
# load the weights
from tensorflow.keras.models import load_model

checkpoint_path = "/home/data/task2/weights.h5"
model= load_model(checkpoint_path)
````
